```python
bronze_schema = dbutils.widgets.get("bronze_schema")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {bronze_schema}")
```

```python
orders_csv_path = "/Volumes/workspace/default/orders"
```

```python
orders_df = (
spark.read.format("csv")
.option("header", True)
.option("inferSchema", True)
.load(orders_csv_path)
)
```

```python
orders_df.write.format("delta").mode("overwrite").saveAsTable(f"{bronze_schema}.orders_bronze")
```
